```python
from functools import cache
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

torch.random.manual_seed(0)
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-128k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-128k-instruct")

messages = [
    {"role": "system", "content": "You are a helpful AI assistant."},
    {"role": "user", "content": "Can you provide ways to eat combinations of bananas and dragonfruits?"},
    {"role": "assistant", "content": "Sure! Here are some ways to eat bananas and dragonfruits together: 1. Banana and dragonfruit smoothie: Blend bananas and dragonfruits together with some milk and honey. 2. Banana and dragonfruit salad: Mix sliced bananas and dragonfruits together with some lemon juice and honey."},
    {"role": "user", "content": "What about solving an 2x + 3 = 7 equation?"},
]

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

generation_args = {
    "max_new_tokens": 500,
    "return_full_text": False,
    "temperature": 0.0,
    "do_sample": False,
}

output = pipe(messages, **generation_args)
print(output[0]['generated_text'])
```

```python

def generate(prompt: str, max_new_tokens: int = 100, temperature: float = 0.7, top_p: float = 0.95, top_k: int = 50):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # If temperature == 0, use deterministic decoding (no sampling)
    do_sample = temperature > 0

    out = model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=do_sample,
        temperature=temperature if do_sample else None,
        top_p=top_p if do_sample else None,
        top_k=top_k if do_sample else None,
        repetition_penalty=1.05,
        eos_token_id=tokenizer.eos_token_id,
        pad_token_id=tokenizer.eos_token_id,
    )

    return tokenizer.decode(out[0], skip_special_tokens=True)
```

```python
prompt = "Explain temperature in LLM decoding in one sentence."

print("=== Temperature 0 (deterministic) ===")
print(generate(prompt, temperature=0))

print("\n=== Temperature 0.7 (balanced) ===")
print(generate(prompt, temperature=0.7))

print("\n=== Temperature 2.0 (more random) ===")
print(generate(prompt, temperature=2.0))
```

```python
prompt = "Explain top_p in LLM decoding in one sentence."

print("=== Temperature 0.7 === top_p = 0.95 === ")
print(generate(prompt, temperature=0.7, top_p = 0.95))

print("\n=== Temperature 0.7 === top_p = 0.2 === ")
print(generate(prompt, temperature=0.7, top_p = 0.2))
```

```python
prompt = "Explain top_k in LLM decoding in one sentence."

print("\n=== Temperature 0.7 === top_k=50 === ")
print(generate(prompt, temperature=0.7, top_k = 50))


print("\n=== Temperature 0.7 === top_k=10 === ")
print(generate(prompt, temperature=0.7, top_k = 10))


print("\n=== Temperature 0.7 === top_k=100 === ")
print(generate(prompt, temperature=0.7, top_k = 100))

```
